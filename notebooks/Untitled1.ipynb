```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from quenched_galaxies.definitions import data
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,8)

x = data['MUFASA']['log_Mstar']
y = data['MUFASA']['log_Mstar'] - data['MUFASA']['log_Mhalo']
c = data['MUFASA']['sSFR_1Gyr'] * 1.0E9

select = c > 0

print np.max(np.log10(c[select])), np.min(np.log10(c[select]))

t = ax.scatter(x[select], y[select],
           c = np.log10(c[select]), alpha = 0.5, vmin = -2, vmax = 0.2)
ax.set_xlim(8.0,12.0)
ax.set_ylim(-3,3)
ax.set_xlabel(r'log(M$_{*}$/M$_{\odot}$)')
ax.set_ylabel(r'log(M$_{*}$/M$_{\rm halo}$)')
fig.savefig('mufasa_mstar_mvir_test.png')
```
